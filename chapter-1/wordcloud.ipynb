```python
import wikipedia
import re
```

```python
#specify the title of wikipedia

wiki = wikipedia.page('Web scraping')
```

```python
#extract the plain text content  of the page

text = wiki.content
```

```python
#clean text

text = re.sub('==.*?==+' , '',  text)
```

```python
text = text.replace('\n','')
```

# WORDCLOUD

```python
import matplotlib.pyplot as plt
```

```python
def plot_cloud(wordcloud):
    
    #set figure size
    plt.figure(figsize=(40,30))
    
    #display image
    
    plt.imshow(wordcloud)
    plt.axis('off')
```

```python
# Import package
from wordcloud import WordCloud, STOPWORDS
```

```python
# Generate word cloud
wordcloud = WordCloud(width = 3000, height = 2000, 
                      random_state=1, background_color='green', 
colormap='rainbow', collocations=False, 
        stopwords = STOPWORDS).generate(text)# Plot
plot_cloud(wordcloud)
```

# USING CUSTOM IMAGES

```python
import numpy as np
from PIL import Image
```

```python
#import image to change it to np array

mask = np.array(Image.open('twitter.png'))
```

```python
#generate worldcloud

wordcloud = WordCloud(width=3000 , height=2000 , 
                     random_state=42 , background_color='navy',
                    colormap='rainbow' , 
                     collocations=False  ,stopwords=STOPWORDS , 
                mask=mask).generate(text)
```

```python
plot_cloud(wordcloud)
```
